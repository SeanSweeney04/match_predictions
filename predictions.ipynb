```python
matches["venue_code"] = matches["Location"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["Opponent"].astype("category").cat.codes
```

```python
matches['Date'] = matches['Date'].astype(str)
```

```python
matches['Hour'] = matches['Date'].str.extract(r'(\d{2})(?=:\d{2})').astype(int)
```

```python
matches[['Date', 'Time']] = matches['Date'].str.split(' ', expand=True)
```

```python
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
matches['day_code'] = matches['Date'].dt.dayofweek
```

```python
matches['Target'] = (matches["Result"] == "W").astype(int)
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["Date"] < "13/12/2023 "]
```

```python
test = matches[matches["Date"] > "13/12/2023 "]
```

```python
predictors = ["venue_code", "Hour", "opp_code", "day_code"]
```

```python
rf.fit(train[predictors], train["Target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["Target"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["Target"], prediction=preds))
```

```python
pd.crosstab(index=combined["actual"], columns = combined["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["Target"], preds)
```

```python
def make_predictions(data, predictors):
    train = data[data["Date"] < '13/12/2023']
    test = data[data["Date"] > '13/12/2023']
    rf.fit(train[predictors], train["Target"])
    preds = rf.predict(test[predictors])
    precision = precision_score(test["Target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches, predictors)
```
